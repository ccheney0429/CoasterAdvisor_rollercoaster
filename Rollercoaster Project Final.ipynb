```python
%load_ext pycodestyle_magic
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import math
import time, datetime
```

Algorithm: (.5*gforce)((displacement/(2h/g*sin**2(theta))**.5)/10)((duration/ max(duration))+(speed/max(speed))+(height/(max(height))))

```python
rollers = pd.read_csv('rollercoaster.csv')
rollers.columns = rollers.columns.str.strip()
rollers.columns
rollers.head()
rollers.at[111, 'Height (feet)'] = 200
rollers[111:112]
rollers['Height (feet)'] = rollers['Height (feet)'].astype(float)
rollers['Length (feet)'] = rollers['Length (feet)'].astype(float)
rollers = rollers.rename(columns={"Height (feet)": "Height"})
rollers[['Name', 'Height']]
rollers = rollers.rename(columns={"Speed (mph)": "Speed"})
rollers[['Name', 'Speed']]
rollers = rollers.rename(columns={"Drop (feet)": "Drop"})
rollers = rollers.rename(columns={"Length (feet)": "Length"})
rollers = rollers.rename(columns={"Duration (min:sec)": "Duration"})
rollers.loc[134, 'Duration'] = '0:00'
rollers.iloc[134]
rollers['Duration'] = rollers['Duration'].fillna('0:00')
rollers['Duration'] = pd.to_datetime(rollers['Duration'], format='%M:%S')
rollers['Duration']


def convertTime(t):
    return int(datetime.timedelta
               (minutes=t.minute, seconds=t.second).total_seconds())


rollers['Duration'] = rollers['Duration'].apply(convertTime)
```

```python
rollers.columns


rollers['G Force'] = pd.to_numeric(rollers['G Force'], errors='coerce')
rollers['Drop'] = pd.to_numeric(rollers['Drop'], errors='coerce')
rollers['Drop'] = rollers['Drop'].fillna(rollers['Drop'].mean())
rollers['G Force'] = rollers['G Force'].fillna(rollers['G Force'].mean())
rollers['Vertical Angle (degrees)'] = rollers['Vertical Angle (degrees)'].transform(lambda x: x.fillna(x.mean()))
rollers['Speed'] = rollers['Speed'].fillna(rollers['Speed'].mean())
rollers['Height'] = rollers['Height'].fillna(rollers['Height'].mean())
rollers['Vertical Angle (degrees)'] = rollers['Vertical Angle (degrees)'].apply(math.radians)


# rollers['Drop']
#rollers['Vertical Angle (degrees)']
# rollers.dtypes
```

```python
rollers['Score'] = (.5 * rollers['G Force'] * ((rollers['Drop'])/(2 * rollers['Drop']/9.8 * (rollers['Vertical Angle (degrees)'].apply(math.sin))**2)**.5)/10)+((rollers['Speed']/rollers.Speed.max())+(rollers['Duration']/rollers.Duration.max())+(rollers['Height']/rollers.Height.max()))
rollers.sort_values(by=['Score'], ascending=False)[0:10]
```
